#*Cereal Packet Weight Analysis *

##A quality control analyst is reviewing the packaging process for a popular breakfast cereal brand. The company claims that each cereal packet contains at least 450 grams of product. A random sample of 7 cereal packets was selected from the production line to investigate this claim.

##The recorded weights (in grams) of the sampled packets are:
##445, 453, 447, 451, 440, 460, 449

###Construct boxplot of the data. Are there any unusual observations in the data set ?

```r
weights<-c(445, 453, 447, 451, 440, 460, 449)
```

```r
weights
```

```r
boxplot(
  weights,
  ylab="Weights(grame)",
  main="Boxplot",
  col="lightgreen"
)
```

###There are no any outliers in the boxplot. So there is no any unusual data points in dataset.

##Based on a normal Q-Q plot, Check parent distribution of weights is normally distributed ?


```r
qqnorm(weights,main="Q-Q plot",col="red")
qqline(weights,col="violet")
```

```r
shapiro.test(weights)
```

### P- values is greater than 0.05 (Alpha values). So dataset is normall.
## So our dataset is following normall assumptions and no unusual datapoints its important to study mean.

### Check Is there any evidence at the 5% significance level that the packets are underweight?

### H0: mu>=450
### H1 mu<450    "Left tail test"

```r
t.test(weights,mu=450,alternative="less")
```

## P-value is greater than 0.05. so we do not reject null hypothesis. therefore we do not have enough evidence to say that packets are underweight.
